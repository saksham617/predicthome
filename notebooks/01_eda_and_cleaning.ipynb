```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/real_estate_main.csv')
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
```

```python
df.head()
```

```python
print("Missing values:")
print(df.isnull().sum())
```

```python
# Drop columns we don't need
cols_to_drop = [
    'Unnamed: 0', 'Name', 'Rate', 'property', 
    'carpet.area', 'facing', 'overlooking', 
    'ownership', 'parking', 'balcony', 
    'city', 'total_area'
]

df = df.drop(columns=cols_to_drop)
print("Shape after dropping:", df.shape)
print("\nRemaining columns:", df.columns.tolist())
```

```python
# Check missing in remaining columns
print("Missing values remaining:")
print(df.isnull().sum())
print("\nData types:")
print(df.dtypes)
```

```python
# Fix missing values

# 1. status - fill with most common value
df['status'] = df['status'].fillna(df['status'].mode()[0])

# 2. floor - fill with most common value  
df['floor'] = df['floor'].fillna(df['floor'].mode()[0])

# 3. transaction - only 1 missing, fill with most common
df['transaction'] = df['transaction'].fillna(df['transaction'].mode()[0])

# 4. bathroom - fill with median
df['bathroom'] = df['bathroom'].fillna(df['bathroom'].median())

# 5. Rate_per_sqft - fill with median
df['Rate_per_sqft'] = df['Rate_per_sqft'].fillna(df['Rate_per_sqft'].median())

# 6. carpet_area_sqft - fill with median
df['carpet_area_sqft'] = df['carpet_area_sqft'].fillna(df['carpet_area_sqft'].median())

# Confirm no missing values left
print("Missing values after fixing:")
print(df.isnull().sum())
print("\nShape:", df.shape)
```

```python
# Check unique values in categorical columns
print("STATUS values:")
print(df['status'].value_counts())

print("\nTRANSACTION values:")
print(df['transaction'].value_counts())

print("\nFLOOR sample values:")
print(df['floor'].value_counts().head(10))

print("\nLOCATION - total unique:")
print(df['location'].nunique())
print("\nTop 10 locations:")
print(df['location'].value_counts().head(10))
```

```python
df['status'] = df['status'].apply(
    lambda x: 'Ready to Move' if x == 'Ready to Move' else 'Under Construction'
)

print(df['status'].value_counts())
```

```python
# Extract floor number from "1 out of 4" format
def extract_floor(floor_str):
    floor_str = str(floor_str).lower().strip()
    if 'ground' in floor_str:
        return 0
    try:
        return int(floor_str.split('out')[0].strip())
    except:
        return 0

df['floor'] = df['floor'].apply(extract_floor)

print("Floor values after cleaning:")
print(df['floor'].value_counts().head(10))
print("\nMin floor:", df['floor'].min())
print("Max floor:", df['floor'].max())
```

```python
# Group rare locations into 'Other'
# Any location with less than 5 listings → 'Other'
location_counts = df['location'].value_counts()
rare_locations = location_counts[location_counts < 5].index

df['location'] = df['location'].apply(
    lambda x: 'Other' if x in rare_locations else x
)

print("Locations after grouping:")
print("Unique locations now:", df['location'].nunique())
print("\nTop 15 locations:")
print(df['location'].value_counts().head(15))
```

```python
# Extract area from Name column
raw_df = pd.read_csv('../data/real_estate_main.csv')

def extract_area(name):
    try:
        # Get last part after final comma
        area = name.strip().split(',')[-1].strip()
        return area
    except:
        return 'Other'

df['area'] = raw_df['Name'].apply(extract_area)

print("Unique areas:")
print(df['area'].value_counts())
```

```python
# Map everything to clean zone names
def clean_area(area):
    area = str(area).strip().lower()
    
    if 'new delhi' in area or 'delhi' in area:
        return 'New Delhi'
    elif 'gurgaon' in area or 'gurugram' in area:
        return 'Gurgaon'
    elif 'greater noida' in area:
        return 'Greater Noida'
    elif 'noida' in area:
        return 'Noida'
    elif 'ghaziabad' in area:
        return 'Ghaziabad'
    elif 'faridabad' in area:
        return 'Faridabad'
    else:
        return 'Other'

df['area'] = df['area'].apply(clean_area)

print("Clean area distribution:")
print(df['area'].value_counts())
print("\nTotal unique areas:", df['area'].nunique())
```

```python
# Drop old location column, keep area
df = df.drop(columns=['location'])

# Check final shape and columns
print("Final dataset:")
print(df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
print("PRICE stats:")
print(df['Price'].describe())

print("\nCARPET AREA stats:")
print(df['carpet_area_sqft'].describe())

print("\nBEDROOM stats:")
print(df['bedroom'].describe())

print("\nBATHROOM stats:")
print(df['bathroom'].describe())
```

```python
print("Before outlier removal:", df.shape)

# 1. Remove impossible carpet areas (less than 200 sqft)
df = df[df['carpet_area_sqft'] >= 200]

# 2. Remove carpet area per bedroom less than 200 sqft
# (no one lives in 100 sqft per room)
df = df[df['carpet_area_sqft'] / df['bedroom'] >= 200]

# 3. Remove extreme price outliers
# Keep properties between 5 lakh and 50 crore
df = df[df['Price'] >= 500000]
df = df[df['Price'] <= 500000000]

# 4. Remove more bathrooms than bedrooms + 2
df = df[df['bathroom'] <= df['bedroom'] + 2]

print("After outlier removal:", df.shape)
print("\nPRICE stats after:")
print(df['Price'].describe())
```

```python
# Convert price to lakhs (easier to understand)
df['Price_Lakhs'] = df['Price'] / 100000

# Drop original Price column
df = df.drop(columns=['Price'])

# YOUR ORIGINAL FEATURE 1:
# Average price per sqft by area
area_avg_ppsf = df.groupby('area')['Rate_per_sqft'].median()
df['area_median_ppsf'] = df['area'].map(area_avg_ppsf)

# YOUR ORIGINAL FEATURE 2:
# How does this property compare to its area's median?
df['ppsf_vs_area'] = df['Rate_per_sqft'] / df['area_median_ppsf']

print("New features added:")
print(df[['area', 'Rate_per_sqft', 'area_median_ppsf', 'ppsf_vs_area']].head(10))
print("\nFinal columns:", df.columns.tolist())
print("Final shape:", df.shape)
```
